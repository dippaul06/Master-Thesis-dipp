```python
## Traversing the CSV file
with open(user_file, newline='') as uf:
    user_data = csv.reader(uf, delimiter=',')
    user_data_list = list(user_data)
```

```python
count = 0
for pos, item in enumerate(user_data_list):
    try:
        data = item[3]
        if(data == ''):
            user_data_list[pos][3] = "['none', 'none', 'none', 'none', 'none']"
        else:
            for pos_loc, item_loc in enumerate(loc_data_list):
                try:
                    data_loc_test = item_loc
                    if(data_loc_test==data):
                        rep = loc_data_list[pos_loc+1]
                        user_data_list[pos][3] = rep
                    else:
                         count += 1
                         continue
                except IndexError:
                    continue
    except IndexError:
                    continue
    
print(count)
```
