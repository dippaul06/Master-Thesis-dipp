```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_url = "/home/dipp/Github/Playground/Data/FOR PLOT/urls.csv"
data_text = "/home/dipp/Github/Playground/Data/FOR PLOT/texts.csv"
data_symbols = "/home/dipp/Github/Playground/Data/FOR PLOT/symbols.csv"
#res = "/home/dipp/Github/Playground/Data/FOR PLOT/Res"
```

```python
url = pd.read_csv(data_url)
for_plot_df = pd.DataFrame(url)
df = for_plot_df.head(50)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.bar(X, Y)
plt.xlabel("URL")
plt.ylabel("Number")
#plt.yscale('symlog')
plt.xticks(rotation='90')
plt.savefig('/home/dipp/Github/Playground/Data/FOR PLOT/Res/URL_bar.jpg',bbox_inches='tight', dpi=150)
plt.show()

for_plot_df_1 = pd.DataFrame(url)
df = for_plot_df_1.head(100)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.plot(X, Y)
plt.xlabel("URL")
plt.ylabel("Number")
#plt.yscale('symlog')
plt.xticks(rotation='90')
plt.savefig('/home/dipp/Github/Playground/Data/FOR PLOT/Res/URL_line.jpg',bbox_inches='tight', dpi=150)
plt.show()
```
