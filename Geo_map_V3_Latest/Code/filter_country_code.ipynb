```python
import csv
import pandas as pd
```

```python
data_file = "/home/dipp/Github/Master-Thesis-dipp/Geo_map_V3_Latest/Data/Location_with_only_country_code_V2_Latest.csv"
```

```python
country_data = {}
with open(data_file, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    count = 0
    for row in reader:
        if(row['i'] != 'None' and row['j'] != 'None' and row['i'] != row['j']):
            souce_loc = row['i'].replace("'","")
            destination_loc = row['j'].replace("'","")

            country_data[count] = [souce_loc, destination_loc, float(row['count'])]
            count+=1
```

```python
country_data
```

```python
df = pd.DataFrame.from_dict(country_data, orient='index', columns=['Source','Destination','Number of Tweets'])
```
