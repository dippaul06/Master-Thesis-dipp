```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import seaborn as sns
```

```python
cols = ['Degree', 'Count']
file_name_all = "/home/dipp/Github/Master-Thesis-dipp/DEGREE-PLOT/degree_all.csv"
file_name_in = "/home/dipp/Github/Master-Thesis-dipp/DEGREE-PLOT/degree_in.csv"
file_name_out = "/home/dipp/Github/Master-Thesis-dipp/DEGREE-PLOT/degree_out.csv"
```

```python
def csv_to_df(file_name):
    with open(file_name, newline='') as f:
        reader = csv.reader(f)
        data = list(b for a, b in reader)
        res = [(i, data.count(i)) for i in set(data)]
        df = pd.DataFrame(res, columns=cols)
    
    return df
```

```python
df_all = csv_to_df(file_name_all)
df_in = csv_to_df(file_name_in)
df_out = csv_to_df(file_name_out)
```
