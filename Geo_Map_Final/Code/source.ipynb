```python
#US → GB, CA, IN, AU, IT, JP, DE, HK, ZA, NL

import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Coordinates for US
us_coordinates = (-100.445882, 39.7837304)

# Coordinates for the 10 countries
countries = {
    'GB': ('-3.2765753', '54.7023545'),  # Great Britain (London)
    'CA': ('-107.991707', '61.0666922'),  # Canada (Toronto)
    'IN': ('78.6677428', '22.3511148'),  # India (New Delhi)
    'AU': ('134.755', '-24.7761086'),  # Australia (Sydney)
    'IT': ('12.674297', '42.6384261'),  # Italy (Rome)
    'JP': ('139.2394179', '36.5748441'),  # Japan (Tokyo)
    'DE': ('10.4478313', '51.1638175'),  # Germany (Berlin)
    'HK': ('114.1849161', '22.350627'),  # Hong Kong
    'ZA': ('24.991639', '-28.8166236'),  # South Africa (Cape Town)
    'NL': ('5.541246849406163', '52.24764975')  # Netherlands (Amsterdam)
}

# Create a GeoDataFrame with the 10 countries
gdf = gpd.GeoDataFrame(
    {'Country': countries.keys()},
    geometry=[Point(float(lon), float(lat)) for lon, lat in countries.values()],
    crs='EPSG:4326'
)

# Create a simple world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
ax = world.plot(figsize=(12, 12), color='lightgray', edgecolor='k')

# Plot the points for the 10 countries
gdf.plot(ax=ax, marker='o', color='red', markersize=10)

# Add arrows from the US to each country
arrowprops = dict(arrowstyle='-|>', color='blue', lw=2)
for idx, row in gdf.iterrows():
    plt.annotate(
        '', xy=row.geometry.coords[0], xytext=us_coordinates,
        arrowprops=arrowprops, annotation_clip=False
    )

# Set plot limits
ax.set_xlim([-180, 180])
ax.set_ylim([-90, 90])

# Show the map
plt.title('Top 10 destination coutries of Tweets made from United States')
plt.savefig('/home/dipp/Github/Master-Thesis-dipp/Geo_Map_Final/Result/top_5_source_US_new.pdf', bbox_inches='tight')
plt.show()
```
