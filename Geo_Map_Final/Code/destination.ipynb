```python
# US → GB, IN, CA, AU, DE, ZA, HK, NG, IT, NL,

import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Coordinates for the United States
us_coordinates = (-100.445882, 39.7837304)

# Coordinates for the 10 source countries
sources = {
    'GB': (-3.2765753, 54.7023545),   # Great Britain (London)
    'IN': (78.6677428, 22.3511148),   # India (New Delhi)
    'CA': (-107.991707, 61.0666922),  # Canada (Toronto)
    'AU': (134.755, -24.7761086),  # Australia (Sydney)
    'DE': (10.4478313, 51.1638175),   # Germany (Berlin)
    'ZA': (24.991639, -28.8166236),  # South Africa (Cape Town)
    'HK': (114.1849161, 22.350627),  # Hong Kong
    'NG': (7.9999721, 9.6000359),     # Nigeria (Abuja)
    'IT': (12.674297, 42.6384261),   # Italy (Rome)
    'NL': (5.541246849406163, 52.24764975)     # Netherlands (Amsterdam)
}

# Create a GeoDataFrame with the 10 source countries
gdf = gpd.GeoDataFrame(
    {'Country': sources.keys()},
    geometry=[Point(float(lon), float(lat)) for lon, lat in sources.values()],
    crs='EPSG:4326'
)

# Create a simple world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
ax = world.plot(figsize=(12, 12), color='lightgray', edgecolor='k')

# Plot the points for the source countries and the United States as the destination
gdf.plot(ax=ax, marker='o', color='red', markersize=10)
plt.annotate('US', xy=us_coordinates, xytext=(us_coordinates[0] + 15, us_coordinates[1]))

# Add arrows from the source countries to the United States
arrowprops = dict(arrowstyle='-|>', color='blue', lw=2)
for idx, row in gdf.iterrows():
    plt.annotate(
        '', xy=row.geometry.coords[0], xytext=us_coordinates,
        arrowprops=arrowprops, annotation_clip=False
    )

# Set plot limits
ax.set_xlim([-180, 180])
ax.set_ylim([-90, 90])

# Show the map
plt.title('Tweets from top 10 Source Countries to the United States')
plt.savefig('/home/dipp/Github/Master-Thesis-dipp/Geo_Map_Final/Result/top_5_destination_US_new.pdf', bbox_inches='tight')
plt.show()
```
