```python
import csv
import pandas as pd
```

```python
data_file = '/home/dipp/Github/Master-Thesis-dipp/Geo_Map_Final/Data/Different_Country.csv'
```

```python
source_country_dict = {}
destination_country_dict = {}
```

```python
with open(data_file, 'r') as data_files:
    raw_data = csv.DictReader(data_files)

    for line in raw_data:
        source = line['Source']
        destination = line['Destination']
        
        source_tweet = source_country_dict.get(source)
        destination_tweet = destination_country_dict.get(destination)

        if(source_tweet != None):
            source_country_dict[source] = float(line['Number of Tweets']) + float(source_tweet)
        if(source_tweet == None):
            source_country_dict[source] = float(line['Number of Tweets'])
        if(destination_tweet != None):
            destination_country_dict[destination] = float(line['Number of Tweets']) + float(destination_tweet)
        if(destination_tweet == None):
            destination_country_dict[destination] = float(line['Number of Tweets'])

        source_tweet = 0
        destination_tweet = 0
        
```

```python
source_country_dict
```

```python
source_df = pd.DataFrame(list(source_country_dict.items()), columns=['country', 'number of tweets'])
destination_df = pd.DataFrame(list(destination_country_dict.items()), columns=['country', 'number of tweets'])
```

```python
source_df = source_df.sort_values(by=['number of tweets'], ascending=False)
destination_df = destination_df.sort_values(by=['number of tweets'], ascending=False)
```
