```python
import pandas as pd
import geopandas
from shapely.geometry import Point
```

```python
data_file = '/home/dipp/Github/Master-Thesis-dipp/GeoGraphic HeatMap/Data/Country_Data_V2_Latest.csv'
result_shapefile = '/home/dipp/Github/Master-Thesis-dipp/GeoGraphic HeatMap/Result/geo-shape.shp'
```

```python
df = pd.read_csv(data_file)
df
```

## Converting this Dataframe to a Shapefile to use in GeoPandas 
https://gis.stackexchange.com/questions/147156/making-shapefile-from-pandas-dataframe

```python
df['geometry'] = df.apply(lambda x: Point((float(x.Longitude), float(x.Latitude))), axis=1)
df
```

```python
df_01 = geopandas.GeoDataFrame(df, geometry='geometry')
```

```python
df_01
```

```python
df_01.to_file(result_shapefile, driver='ESRI Shapefile')
```
